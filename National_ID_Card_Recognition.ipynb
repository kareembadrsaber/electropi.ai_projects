```python
# Define the row bounds for slicing
start_row = int(img.shape[0] * 0.5)  # Start from the middle of the image
end_row = int(img.shape[0] * 0.75)   # End three-quarters down the image

# Ensure the row bounds are within the image dimensions
start_row = max(0, start_row)
end_row = min(img.shape[0], end_row)

# Slice the image within bounds
sliced_img = img[start_row:end_row, :]

# Check if the sliced image is not empty before displaying it
if not np.any(sliced_img):
    print("Error: Sliced image is empty.")
else:
    # Show the sliced image
    plt.imshow(cv2.cvtColor(sliced_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # Perform OCR on the sliced image
    extracted_text = pytesseract.image_to_string(sliced_img, lang='ara')
    print(extracted_text)
```
